# Exploration: tune preprocessing parameters
Use this notebook to visually check grayscale/blur/threshold output on real images before locking values into `config/settings.py`.

```python
import sys
sys.path.append('..')

import cv2
import matplotlib.pyplot as plt
from src.preprocessing import to_grayscale, apply_blur, binarize
from src.topology import analyze_topology
from functions.image_io import load_image
```

```python
img = load_image('../data/raw/good/example.jpg')
gray = to_grayscale(img)
blurred = apply_blur(gray)
thresh = binarize(blurred)

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); axes[0].set_title('Original')
axes[1].imshow(gray, cmap='gray'); axes[1].set_title('Grayscale')
axes[2].imshow(blurred, cmap='gray'); axes[2].set_title('Blurred')
axes[3].imshow(thresh, cmap='gray'); axes[3].set_title('Threshold')
for ax in axes: ax.axis('off')
plt.show()
```

```python
contour, hull, defects = analyze_topology(thresh)
print('Defects found:', 0 if defects is None else defects.shape[0])
if defects is not None:
    for i in range(defects.shape[0]):
        s, e, f, d = defects[i, 0]
        print(f'  defect {i}: distance = {d/256.0:.2f} px')
```
